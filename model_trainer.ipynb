```python
from roboflow import Roboflow

# Pega tu clave de API privada de https://app.roboflow.com/settings/api
rf = Roboflow(api_key=f"{input('Ingrese su clave de API privada de Roboflow (https://app.roboflow.com/settings/api)')}")

try:
    # balon-person-yolo-nlffk
    project = rf.project("balon-person-yolo-nlffk")
    dataset = project.version(1).download("yolov8")

    print("\n✅ ¡Dataset descargado con éxito!")
    print(f"El dataset está en: {dataset.location}")

except Exception as e:
    print("\n❌ ¡Error al descargar!")
    print("Por favor, asegúrate de haber reemplazado 'TU_API_KEY_AQUI' con tu clave de API real de Roboflow.")
    print(f"Error original: {e}")
```
